```python
import numpy as np
from scipy import ndimage, misc
import matplotlib.pyplot as plt
from functools import partial
```

```python
imshow = partial(plt.imshow, cmap='gray')
```

```python
im = misc.face()[:, :, 0]
imshow(im);
im.shape
```

```python
768 / 20,  1024 / 20
```

```python
scale = 0.05
# im10 = ndimage.zoom(im, scale, order=1)
im10 = im[::int(1/scale), ::int(1/scale)]
imshow(im10);
assert im10.shape[0] == np.round(im.shape[0] * scale), f"{(im10.shape, im.shape)}"
assert im10.shape[1] == np.round(im.shape[1] * scale), f"{(im10.shape, im.shape)}"
(im10.shape, im.shape)
```

```python
im.mean(), ndimage.zoom(im, 0.1, order=3).mean()
```

```python
im.std(), ndimage.zoom(im, 0.1, order=3).std()
```

```python
scale = 0.05
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)
im1 = ndimage.zoom(im, scale, order=1)
im2 = im[::int(1/scale), ::int(1/scale)]
ax1.imshow(im1);
ax2.imshow(im2);
ax3.imshow(im2[1:, 1:] - im1);
```
